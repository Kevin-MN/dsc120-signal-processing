```python
import numpy as np
import pywt
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  
from matplotlib import cm
from matplotlib import rc
```

```python
img1 = matplotlib.pyplot.imread('image1.png')
```

```python
fig = plt.figure(figsize =(8,4))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray', aspect='equal')  # imshow is how we plot images, cmap is colormap
plt.colorbar()
```

```python
img2 = matplotlib.pyplot.imread('image2.png')
```

```python
fig = plt.figure(figsize =(8,4))
plt.subplot(1, 2, 1)
plt.imshow(img2, cmap='gray', aspect='equal')  # imshow is how we plot images, cmap is colormap
plt.colorbar()
```

```python
# Only do decomposition once
level=5

# 2D Haar
coeffs = pywt.wavedec2(img1, 'haar', level=level)

# Turn into array and back
arr, coeff_slices = pywt.coeffs_to_array(coeffs)
coeffs_reconstruct = pywt.array_to_coeffs(arr, coeff_slices, output_format = 'wavedec2')

print(arr.shape)

plt.figure(figsize=(24,8))
plt.subplot(1,2,1)
plt.imshow(arr,vmax=10)
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(np.abs(arr),vmax=0.1)
plt.colorbar()
plt.show()
```
