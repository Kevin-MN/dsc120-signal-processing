```python
M = 164 # number of rows
N = 220  # number of columns 

x = np.arange(0,N)  # horizontal grid 
y = np.arange(0,M)  # vertical grid 
xx, yy = np.meshgrid(x, y)
```

```python
img1 = matplotlib.pyplot.imread('vertNoise.png')
fig = plt.figure(figsize =(8,4))
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray', aspect='equal')  # imshow is how we plot images, cmap is colormap
plt.colorbar()
```

```python
z_hat = np.fft.fft2(img1) # calculate the 2D DFT in numpy
```

```python
fig = plt.figure(figsize =(20,20))
plt.subplot(1, 3, 1)
plt.imshow(img1,cmap='gray', aspect='equal')
plt.title('$img1$')


plt.subplot(1, 3, 3)
plt.imshow(np.log((np.fft.fftshift(abs(z_hat)))),cmap='gray', aspect='equal',origin='lower')  # don't forget to fftshift
#  For two-dimensional input, swaps first and third quadrants, and second and fourth quadrants.
plt.xticks(ticks=[0,220/2, 220-1], labels=[-220/2,0,220/2]) 
plt.yticks(ticks=[0,164/2, 164-1], labels=[-164/2,0,164/2]) 
plt.title('$dft img1$')
plt.show()
```

```python
z_hat[(xx<-100) & (xx>100)] = 0

cleaned1 = np.abs(np.fft.ifft2(z_hat))
```

```python
fig = plt.figure(figsize =(20,20))
plt.subplot(1, 3, 1)
plt.imshow(cleaned1,cmap='gray', aspect='equal')
plt.title('$img1$')
```
