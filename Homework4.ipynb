```python
def recur_fft(x):
    N = len(x)
    
    if N == 1: #base case single element
        return x
    else:
        x = fft2(x[::2]) + fft2(x[1::2])# unwind and add element-wise
        for k in range(N//2):  #
            xk = x[k]
            x[k] = xk + np.exp(-2*np.pi*1j/N)**k*x[k+N//2] # add scaled exponential 
            x[k+N//2] = xk - np.exp(-2*np.pi*1j/N)**k*x[k+N//2]
        return x
```
